Supongamos que una partícula de masa m está confinada a una caja unidimensional de ancho $L$ La energía potencial para esta situación, que comúnmente se conoce como pozo de potencial infinito. La energía potencial $V(x)$ se se puede expresar como:

$$
V(x)=0\hspace{1cm} 0<x<L,
$$

$$
V(x)=\infty\hspace{1cm} x\leq 0~~y~~x\geq L.
$$

La ecuación de Schrödinger dependiente del tiempo se puede escribir así:


$$
\left[-\frac{\hbar^2}{2m}\frac{\partial^2}{\partial x^2} +V(x) \right]\Psi(x,t)=i\hbar\frac{\partial \Psi(x,t)}{\partial t}
$$

Para el caso que no hay potencial dentro de la caja

$$
-\frac{\hbar^2}{2m}\frac{\partial^2}{\partial x^2}\Psi(x,t)=i\hbar\frac{\partial \Psi(x,t)}{\partial t}
$$

Consideremos un electrón está en una caja de longitud $L=10\times10^{-8}m$ y supongamos que el tiempo $t=0$ la función de onda del electrón tiene la forma:

$$
\Psi(x,0)=\text{exp}\left[-\frac{(x-x_0)^2}{2\sigma^2} \right]e^{ikx},
$$

donde $x_0=L/2, \sigma=1\times10^{-10}m$ y $k=5\times10^{10}m^{-1}$.

Y $\Psi=0$ en la paredes de la caja en $x=0$ y $x=L$.

\
$$
-\frac{\hbar^2}{2m}\frac{\partial^2}{\partial x^2} \Psi(x,t)=i\hbar\frac{\partial \Psi(x,t)}{\partial t}
$$
\
$$
-\frac{\hbar^2}{2m}\frac{\partial^2 \Psi(x,t)}{\partial x^2} =i\hbar\frac{\partial \Psi(x,t)}{\partial t}
$$
\
Dividimos a ambos lados por $i\hbar$
\
$$
-\frac{1}{i\hbar}\frac{\hbar^2}{2m}\frac{\partial^2 \Psi(x,t)}{\partial x^2} =\frac{\partial \Psi(x,t)}{\partial t}
$$
\
$$
\frac{\partial \Psi(x,t)}{\partial t}=\frac{i\hbar}{2m}\frac{\partial^{2}\Psi(x,t)}{\partial x^{2}}\hspace{1cm}(1)
$$
\
Por medio de diferencias finitas se puede escribir la segunda derivada de la siguiente manera:


$$
\frac{\partial^{2}\Psi(x,t)}{\partial x^{2}}=\frac{1}{a^{2}}\left[\Psi\left(x+a,t\right)-2\Psi\left(x,t\right)+\Psi\left(x-a,t\right)\right]
$$
\
Por metodo de Euler se tiene que:
\
$$
\Psi\left(x,t+h\right)=\Psi(x,t)+h\frac{\partial\Psi(x,t)}{\partial t}
$$
\
Donde $a$ y $h$ son los  pasos de separacion espacial y temporal respectivamente.

Reemplazando en la ecuación $(1)$ en el método de Euler

$$
\Psi(x,t+h)=\Psi(x,t)+h\left[\frac{i\hbar}{2m} \frac{\partial ^2 \Psi(x,t)}{\partial x^2}\right]
$$

$$
\Psi(x,t+h)=\Psi(x,t)+\frac{ih\hbar}{2m} \frac{\partial ^2 \Psi(x,t)}{\partial x^2}
$$

Reemplazamos la segunda derivada

$$
\Psi\left(x,t+h\right)=\Psi\left(x,t\right)+\frac{ih\hbar}{2m}\left[\frac{1}{a^2}\left[\Psi\left(x+a,t\right)+\Psi\left(x-a,t\right)-2\Psi\left(x,t\right)\right]\right]\hspace{1cm}(2)
$$


Sí hacemos $h\to-h$ y $t\rightarrow t+h$


$$
\Psi\left(x,t\right)=\Psi\left(x,t+h\right)-\frac{ih\hbar}{2m a^{2}}\left[\Psi\left(x+a,t+h\right)+\Psi\left(x-a,t+h\right)-2\Psi\left(x,t+h\right)\right]\hspace{1cm}(3)
$$

Sí restamos las ecuaciones $(3)$ y $(2)$

$$
\Psi(x,t)-\Psi(x,t+h)=\Psi\left(x,t+h\right)-\frac{ih\hbar}{2m a^{2}}\left[\Psi\left(x+a,t+h\right)+\Psi\left(x-a,t+h\right)-2\Psi\left(x,t+h\right)\right]-\Psi\left(x,t\right)-\frac{ih\hbar}{2ma^2}\left[\Psi\left(x+a,t\right)+\Psi\left(x-a,t\right)-2\Psi\left(x,t\right)\right]
$$

Reorganizando para $t+h$ y $t$ a cada lado de la igualdad, se obtiene el metodo de Crank-Nicolson.

$$
-\Psi(x,t+h)-\Psi\left(x,t+h\right)+\frac{ih\hbar}{2m a^{2}}\left[\Psi\left(x+a,t+h\right)+\Psi\left(x-a,t+h\right)-2\Psi\left(x,t+h\right)\right]=-\Psi(x,t)-\Psi\left(x,t\right)-\frac{ih\hbar}{2ma^2}\left[\Psi\left(x+a,t\right)+\Psi\left(x-a,t\right)-2\Psi\left(x,t\right)\right]
$$

$$
-2\Psi(x,t+h)+\frac{ih\hbar}{2m a^{2}}\left[\Psi\left(x+a,t+h\right)+\Psi\left(x-a,t+h\right)-2\Psi\left(x,t+h\right)\right]=-2\Psi(x,t)-\frac{ih\hbar}{2ma^2}\left[\Psi\left(x+a,t\right)+\Psi\left(x-a,t\right)-2\Psi\left(x,t\right)\right]
$$


Multiplicando a ambos lados de la ecuación por $(-1)$

$$
2\Psi(x,t+h)-\frac{ih\hbar}{2m a^{2}}\left[\Psi\left(x+a,t+h\right)+\Psi\left(x-a,t+h\right)-2\Psi\left(x,t+h\right)\right]=2\Psi(x,t)+\frac{ih\hbar}{2ma^2}\left[\Psi\left(x+a,t\right)+\Psi\left(x-a,t\right)-2\Psi\left(x,t\right)\right]
$$

Dividimos a ambos lados de la ecuación anterior por $2$

$$
\Psi\left(x,t+h\right)-\frac{ih\hbar}{4m a^{2}}\left[\Psi\left(x+a,t+h\right)+\Psi\left(x-a,t+h\right)-2\psi\left(x,t+h\right)\right]=\\
\Psi\left(x,t\right)+\frac{ih\hbar}{4m a^{2}}\left[\Psi\left(x+a,t\right)+\Psi\left(x-a,t\right)-2\Psi\left(x,t\right)\right]
$$


Podemos reescribir de la siguiente manera:


$$
\left(-\frac{ih\hbar}{4m a^{2}}\right)\Psi\left(x+a,t+h\right)+\left(1+\frac{ih\hbar}{2m a^{2}}\right)\Psi\left(x,t+h\right)+\left(-\frac{ih\hbar}{4m a^{2}}\right)\Psi\left(x-a,t+h\right)=
\left(\frac{ih\hbar}{4m a^{2}}\right)\Psi\left(x+a,t\right)+\left(1-\frac{ih\hbar}{2m a^{2}}\right)\Psi\left(x,t\right)+\left(\frac{ih\hbar}{4m a^{2}}\right)\Psi\left(x-a,t\right)
$$

Las ecuaciones de Crank-Nicolson se pueden reescribir de la forma:


$$
\vec{A} \Psi\left(t+h\right)=\vec{B} \Psi\left(t\right)
$$

Donde las matrices $\vec{A}$ y $\vec{B}$ son simetricas y tridiagonales

$$
\vec{A}=\begin{pmatrix}
a_1 & a_2 &     &   \\
a_2 & a_1 & a_2 &   \\
  & a_2 & a_1 & a_2\\
  &     & a_2 & a_1 & \\
  &     & &&\ddots
  \end{pmatrix},\hspace{2cm} \vec{B}=\begin{pmatrix}
b_1 & b_2 &     &   \\
b_2 & b_1 & b_2 &   \\
  & b_2 & b_1 & b_2\\
  &     & b_2 & b_1 & \\
  &     & &&\ddots
  \end{pmatrix}
$$

con 

$$
a_1=1+\frac{ih\hbar}{2m a^{2}},\hspace{1cm} a_2=-\frac{ih\hbar}{4m a^{2}},\hspace{1cm} b_1=1-\frac{ih\hbar}{2m a^{2}},\hspace{1cm} b_2=\frac{ih\hbar}{4m a^{2}}
$$

```python
from scipy.linalg import solve_banded
import numpy as np
from scipy.constants import m_e,hbar
from matplotlib import animation
import matplotlib.pyplot as plt
```

```python
#Parametros
L=1e-8  #Longitud de la caja
N=1000  #Cortes espaciales
a=L/N   #Separación espacial
h=1e-18 #Separación temporal
x0=L/2
sigma=1e-10
k=5e10

#Posiciones a traves de la caja
x=np.linspace(0,L,N+1)

#Función de onda inicial
psi=np.zeros(N+1,complex)
psi[1:N]=np.exp(-((x[1:N]-x0)**2)/(2*sigma**2))*np.exp(1j*k*x[1:N])

#Elementos matriciales
a1=1+((h*1j*hbar)/(2*m_e*(a**2)))
a2=(-1)*((h*1j*hbar)/(4*m_e*(a**2)))
b1=1-(1j*h*hbar)/(2*m_e*a**2)
b2=(1j*h*hbar)/(4*m_e*a**2)

#Matriz A
A=np.zeros((3,N+1),complex)

#Banda de diagonales de la matriz A
A[0,:]=a2
A[1,:]=a1
A[2,:]=a2

t=0
t_final=1e-15
v=np.zeros(N+1,complex)
solucion=[]

while t<t_final:
    v[1:N]=b1*psi[1:N]+b2*(psi[2:N+1]+psi[0:N-1])
    psi=solve_banded((1,1),A,v)
    solucion.append(psi) 
    t+=h  
    
sol=np.array(solucion)
solReal=sol.real
```
